```python
# Eliminar las columnas con informacion duplicada o innecesarias

columnas_a_eliminar = ['AAAA', 'MM', 'DD', 'HH', 'Altura', 'Dirección Normalizada', "Cruce", "XY (CABA)", "Calle"]
df = df.drop(columns=columnas_a_eliminar)
```

```python
df.head(3)
```

```python
# Guardar el DataFrame modificado en un nuevo archivo Excel
archivo_ETL = r'D:\user\Documents\SANDRA 2023\HENRY\PROYECTO 2\ETL\homicidios_ETL.xlsx'
df.to_excel(archivo_ETL, index=False)
```

```python
# adecuo la columna fecha solo a año, mes y dia sin hora

# Convertir la columna 'FECHA' a tipo datetime
df['FECHA'] = pd.to_datetime(df['FECHA'])

# Extraer solo la parte de fecha (año, mes y día)
df['FECHA'] = df['FECHA'].dt.date
```

```python
# Guardar el DataFrame modificado en un nuevo archivo Excel
df.to_excel(archivo_ETL, index=False)
```

```python
# Reemplazar "GRAL PAZ" por "Avenida" en la columna "TIPO_DE_CALLE" ya que en la distribucion de victimas por calle es significante la cantidad de victimas
# en GRAL PAZ y en realidad es una AVENIDA
df['TIPO_DE_CALLE'] = df['TIPO_DE_CALLE'].replace('GRAL PAZ', 'AVENIDA')
```

```python
# Reemplazar cualquier ocurrencia de "AVENIDA" por "Avenida" (mayúsculas o minúsculas)
df['TIPO_DE_CALLE'] = df['TIPO_DE_CALLE'].str.upper().replace('Avenida', 'AVENIDA')
```
